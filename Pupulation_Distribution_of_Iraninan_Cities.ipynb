#import Libraries and Data

## import labraries

```
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

##import data

```
df = pd.read_csv('cities.csv')
print(df)
```

##Data information

```
print(type(df))
print(df.shape)
print(df.size)
df.info()
print(f'columns_names = {df.columns}')
print(f'columns_mean = {df['Population'].mean()}')
print(f'columns_var = {df['Population'].var()}')
print(f'columns_std = {df['Population'].std()}')
```

#Depicting map base on latitude and longitude

```
plt.figure(figsize=(12,8))
m=Basemap(
    projection='merc',
    llcrnrlon=44,
    llcrnrlat=25,
    urcrnrlon=64,
    urcrnrlat=40,
    resolution='i'
)
m.drawcoastlines()
m.drawcountries()
x, y = m(df["Longitude"].values,df["Latitude"].values)

m.scatter(x, y, s=df['Population']/10000, c=df['Population'], cmap="plasma_r" , alpha=0.8)
for i in range(len(df)):
    plt.text(x[i], y[i], df['citiesname'][i], fontsize=8)

plt.colorbar(label="Population")
plt.title(" Pupulation Distribution of Iraninan Cities")
plt.show()
```
